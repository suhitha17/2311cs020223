```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('pharma_data.csv')

df.drop_duplicates(inplace=True)
df.isnull().sum()

sns.barplot(x='Region', y='Sales', data=df, estimator=sum)
plt.title('Total Sales per Region')
plt.show()

sns.scatterplot(x='Marketing_Spend', y='Sales', data=df)
plt.title('Marketing Spend vs Sales')
plt.show()

sns.boxplot(x='Age_Group', y='Effectiveness', data=df)
plt.title('Drug Effectiveness by Age Group')
plt.show()

sns.lineplot(x='Trial_Period', y='Sales', hue='Product', data=df, marker='o')
plt.title('Sales Trend by Product over Trial Periods')
plt.show()
sns.heatmap(df[['Sales', 'Marketing_Spend', 'Effectiveness']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
