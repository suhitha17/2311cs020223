```python
#Data visualization
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('banking_data.csv')

df.groupby('Account_Type')['Transaction_Amount'].sum().plot(kind='bar', title='Total Transaction Amount per Account Type')
plt.show()

df['Branch'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Transactions per Branch', ylabel='')
plt.show()
```
